```python
data=[1000,2000,5000,9000,10000]
l=['India','US','Swedan','Germany','china']
e=[0.1,0,0,0,0]
colors=['blue','green','yellow','black','red']
plt.pie(data,labels=l,explode=e,shadow=True,startangle=90,autopct="%1.1f%%",wedgeprops={'edgecolor':'black','linewidth':2},colors=colors)
plt.title("population")
plt.show()
```
